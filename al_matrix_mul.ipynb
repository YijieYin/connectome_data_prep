```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# first read from info 
info = ss.Table('info', base='main')
infodf = info.to_frame()
infodf['root_id'] = infodf.root_id.astype(np.int64)
infodf.head()
```

```python
# also read all the connectivity from syn_proof_analysis
ff = pd.read_feather('/Users/yijieyin/Downloads/syn_proof_analysis_filtered_530.feather')
ff.head()
```

```python
# which neuropils are there? 
ff['neuropil'].unique()
```

```python
# connections in the neuropils we want 
npff = ff[(ff['neuropil'].isin(['AL_L','AL_R'])) & (ff['syn_count'] > 3)]
npff
```

```python
# calculate total number of postsynapses per neuron for those in the neuropil of choice
npids = npff['pre_pt_root_id'].append(npff['post_pt_root_id']).unique()
npids = ff[ff['post_pt_root_id'].isin(npids)].groupby(['post_pt_root_id']).agg(totpostsyn = ('syn_count','sum'),).reset_index()
# add meta data 
npids = pd.merge(npids, infodf[['root_id','cell_type','cell_class','side']], 
                 left_on = 'post_pt_root_id', 
                 right_on = 'root_id')
npids
```

```python
# what are the cell classes? 
npids.cell_class.unique()
```

```python
# group by cell_type or ids 
npids['newtype'] = [npids.loc[i, 'cell_type'] if npids.loc[i, 'cell_type'] != None else npids.loc[i, 'root_id'].astype(str) for i in range(npids.shape[0])]
nptypes = npids.groupby(['newtype']).agg(totpostsyntype = ('totpostsyn', 'sum'),).reset_index()
nptypes
```

```python
# add meta data from info, and total postsynapse number for each neuron 
# default is inner join: discard connections with cells not present in info 
# presynaptic id 
npff = pd.merge(npff, infodf[['root_id','cell_type','cell_class','side']], 
             left_on = 'pre_pt_root_id', 
             right_on = 'root_id')
# add total post synapse and meta info for postsynaptic id 
npff = pd.merge(npff, npids, on = 'post_pt_root_id', suffixes = ('_pre','_post'))
# total post synapse per postsynaptic newtype 
npff = pd.merge(npff, nptypes, on = 'newtype', how = 'left')
# add newtype info for presynaptic neuron 
npff['newtype_pre'] = [npff.loc[i, 'cell_type_pre'] if npff.loc[i, 'cell_type_pre'] != None else npff.loc[i, 'pre_pt_root_id'].astype(str) for i in range(npff.shape[0])]
# calculate neuron-neuron input proportion 
npff['inprop'] = npff.syn_count / npff.totpostsyn
npff
```

```python
# type-type input proportion 
ttnpff = npff.groupby(['newtype_pre','newtype','totpostsyntype','cell_class_pre','cell_class_post']).agg(typesyncount = ('syn_count', 'sum'),).reset_index()
ttnpff['inproptype'] = ttnpff['typesyncount'] / ttnpff['totpostsyntype']
ttnpff.sort_values(by = 'inproptype', ascending = False)
```

```python
# check some by writing ids to clipboard 
npids.loc[npids.cell_type == 'l2LN23', 'root_id'].to_clipboard(index=False,header=False)
```

# Matrix making and multiplication

```python
typeclassall = pd.concat([ttnpff[['newtype_pre','cell_class_pre']].rename(columns={'newtype_pre': 'type', 'cell_class_pre': 'cell_class'}), 
                          ttnpff[['newtype','cell_class_post']].rename(columns={'newtype': 'type', 'cell_class_post': 'cell_class'})])
typeclassall.drop_duplicates(inplace=True, ignore_index=True)
typeclassall['idx'] = range(typeclassall.shape[0])
typeclassall
```

```python
# Create a numpy array with the maximum number of nodes
max_node = typeclassall.shape[0]
adj_matrix = np.zeros((max_node, max_node))

# Loop through the edges and add the weight to the appropriate cells in the adjacency matrix
for _, edge in ttnpff.iterrows():
    adj_matrix[typeclassall[typeclassall.type == edge['newtype']].idx, typeclassall[typeclassall.type == edge['newtype_pre']].idx] = edge['inproptype']

adj_matrix
```

```python
adj_matrix.shape
```

```python
typeclassall.cell_class.unique()
```

```python
# now, make the starting matrix, which is identity for the sensory neurons only 
im = np.zeros((max_node, max_node))

for i in range(typeclassall.shape[0]):
    if typeclassall.loc[i, 'cell_class'] == 'olfactory':
        im[i,i] = 1

im
```

```python
# now, matrix multiplication 
steps = 5 
data = []

data.append(adj_matrix@im)

for step in range(steps-2): 
    
    data.append(np.linalg.matrix_power(adj_matrix, step + 2) @ im)
data
```
